```python
import pandas as pd
```

```python
#Tree info
A = ['a1', 'a2', 'a3']
B = ['b1', 'b2']
C = ['c1', 'c2']
D = ['d1']
E = ['e1', 'e2']

conditions = {'a1':'C', 'a2':'D', 'a3':'B',
            'b1':'E', 'b2':'L1',
            'c1':'L2', 'c2':'L3',
            'd1':'L4',
            'e1':'L5', 'e2':'L6'}

root = 'A'
last_root = root
```

```python
all_route_list = list()
tree_dict = {'A':A}
route = list()
tree_flag = True
while tree_flag:
    
    print(f'Tree Dict before of a period: {tree_dict}')
    
    # Forward
    print('Forward *******************************************')
    forward_flag = True
    while forward_flag:
        
        # Last Root
        last_root = list(tree_dict.keys())[-1]
        last_root_conditions = tree_dict[last_root]
        print(f'Last Root: {last_root}')
        
        # Save the root
        route.append(last_root)
        print(f'Route: {route}')
        
        condition_list = last_root_conditions
        print(f'Condition List: {condition_list}')
        condition_selected = condition_list[-1] # use last condition
        
        # Save the condition
        route.append(condition_selected)
        print(f'Route: {route}')
        
        # Branch or Leaf
        leaf = conditions[condition_list[-1]]
        if leaf[0] == 'L': # Leaf
            print('***** find a Leaf *****')
            route.append(leaf)
            print(f'Route: {route}')
            all_route_list.append(route.copy())
            forward_flag = False
        else: #Branch
            
            # update tree info
            new_root = conditions[condition_list[-1]]
            tree_dict[new_root] = eval(new_root)
            
            # Go to new root
            last_root = new_root
    print('***************************************************')
    print(f'All routes: {all_route_list}')
    
    print(f'Tree Dict after of the forward: {tree_dict}')
    
    # Backward
    print('Backward ******************************************')
    backward_flag = True
    while backward_flag:
        
        # Select last root
        tree_last_root = list(tree_dict.keys())[-1]
        
        # pop condition of root
        tree_dict[tree_last_root].pop()
        
        # condition_list of this root is empty
        while len(tree_dict[tree_last_root]) == 0:
            # The last root was checked and then we have to go back
            tree_dict.pop(tree_last_root)
            
            # Test the tree for the possibility to going back
            if len(tree_dict.keys()) == 0:
                # Finish Tree
                #print('Finished')
                tree_flag = False
                break
        
            # changed the last root
            tree_last_root = list(tree_dict.keys())[-1]
            
            # pop condition of root
            tree_dict[tree_last_root].pop()
        
            #last_root = route[-2]
            route.pop()
            route.pop()
            #route.pop()
        
        route.pop()
        route.pop()
        route.pop()
        
        print(f'Route: {route}')
        
        backward_flag = False
    print('***************************************************')

print()
print('**************************************************')
print('Finished')
print(f'All routes: {all_route_list}')
```

```python
# print all laws
for i, route in enumerate(all_route_list):
    print(f'{i+1}: {route}')
```

```python
# This function searches the list of conditions in the dataset in order.
# df must be laws and condition_dict must be dataset
# we neet to find datasets laws
# return: (Laws that match to dataset, Last condition that is True, True: If all data conditions are accepted else:False)
def search_in_dataframe(df, condition_dict):
    new_df = df.copy()
    
    all_index = list()
    true_index = list()
    last_allowed_condition = None
    for condition in condition_dict.keys():
        #print(f'Condition: {condition}|{condition_dict[condition]}')
        
        for index in new_df.index:
            try:
                if (condition_dict[condition] in df[condition].values):
                    if df[condition].values[index]==condition_dict[condition]:
                        last_allowed_condition = condition
                        true_index.append(index)
                else:
                    return (new_df, last_allowed_condition, False)
            except:
                return (new_df, last_allowed_condition, False)
            
        new_df = new_df.loc[true_index,:].copy()
        true_index = list()
        #print(new_df)
        
    return (new_df, last_allowed_condition, True)
        
laws = pd.DataFrame([['a2', 'b3', 'c1'], ['a2', 'b1', 'c1'], ['a1', 'b3', 'c1', 'e1']], columns=['A', 'B', 'C', 'E'])
dataset_conditions = {'A':'a1', 'B':'b3', 'C':'c1', 'D':'d1'}
result = search_in_dataframe(laws, dataset_conditions)

print(f'Laws: \n{result[0]}')
print(f'Last condition that is Ture: {result[1]}')
print(f'Is the search finished? {result[2]}')
```

```python
def predict(laws, data_df):
    root = laws.values[0,0]
    result_list = list()
    for index in data_df.index:
        new_laws = laws.copy()
        data = data_df.values[index]
        print(f'{index}-searching conditions: {data}')
        
        #search
        last_root = root
        coulmn_last_root_num_in_law = 0 # for Laws
        coulmn_last_condition_num_in_law = 1 # for Laws
        #print(f'last root {last_root}')
        next_condition = data_df.loc[index, root]
        #print(f'next condition: {next_condition}')
        
        resume_flag = True
        while resume_flag:
            # search in laws
            index_accepted_list = list()
            for i in new_laws.index:
                if new_laws.loc[i,coulmn_last_condition_num_in_law] == next_condition:
                    index_accepted_list.append(i)
                    
            if index_accepted_list == []:
                resume_flag = False
                break
            
            if new_laws.shape[0] != 0:
                new_laws = new_laws.loc[index_accepted_list,:]
            else:
                result_list.append({index:list(new_laws.index)})
                # finish
                
            coulmn_last_root_num_in_law += 2
            coulmn_last_condition_num_in_law += 2
            
            if new_laws.iloc[0, coulmn_last_condition_num_in_law] == None:
                label = new_laws.iloc[0, coulmn_last_root_num_in_law]
                print(f'Label: {label}')
                result_list.append({index:list(new_laws.index)+[label]})
                
                resume_flag = False
            else:
                try:
                    last_root = new_laws.iloc[0, coulmn_last_root_num_in_law]
                    next_condition = data_df.loc[index, last_root]
                except: # there aren't this root in the dataset
                    print(f'Label: {None}')
                    result_list.append({index:list(new_laws.index)+[None]})
                    
    return result_list
              
            
```
